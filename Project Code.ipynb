```python
cols=data.columns
cols=list(cols)

data['Alley'].isnull().sum()
        
```

```python
notuseful = []
for i in cols:
    if data[i].isnull().sum()>0:
        notuseful.append(i)
       
```

```python
notuseful
```

```python
colset,notuseful = set(cols),set(notuseful)
```

```python
colset.difference_update(notuseful)
```

```python
len(colset)
```

```python
import numpy as np
newDataFrame = pd.DataFrame()
for i in colset:
    try:
        newDataFrame[i] = np.array(data[i],dtype=np.float64)
    except:
        setcol=set(data[i])
        d={}
        for n,j in enumerate(setcol):
            d[j]=n
        l=list(data[i])
        newDataFrame[i]=[d[i] for i in l]
```
